```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
```

```python
# Working proerly, just need to input all the username in the field manually as it is the algorithm, 
# You can download an excel file from the Kaggle website and use it.


# Instagram credentials
# Enter your username and password

USERNAME = "username"
PASSWORD = "password"



# List of usernames to follow
# I used some random usernames which were downloaded from the Kaggle website

usernames_to_follow = ['doina''donnahaymagazine', 'dotzsoh', 'emwng', 'enjoyphoenix', 'euvictornogueira', 'evakosmasflores', 'faktastisch', 'fantastic_earth', 'mensfashions', 'gianmarcovalenza', 'giizeleoliveira', 'rayniyel', 'samevanslife', 'sarahjanewai', 'staysaltyflorida', 'stefaniegiesinger']


def login(driver, username, password):
    driver.get("https://www.instagram.com/accounts/login/")
    time.sleep(2)
    
    # Enter username and password
    driver.find_element(By.NAME, "username").send_keys(username)
    driver.find_element(By.NAME, "password").send_keys(password)
    time.sleep(1)
    
    # Log in
    driver.find_element(By.XPATH, "//button[@type='submit']").click()
    time.sleep(5)

def follow_users(driver, users):
    for user in users:
        driver.get(f"https://www.instagram.com/{user}/")
        time.sleep(2)
        
        try:
            follow_button = driver.find_element(By.XPATH, "//button[text()='Follow']")
            follow_button.click()
            print(f"Followed {user}")
        except Exception as e:
            print(f"Could not follow {user}: {e}")
        time.sleep(3)  # Wait between actions to avoid detection

def main():
    # Set up WebDriver
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
    driver.maximize_window()
    
    try:
        login(driver, USERNAME, PASSWORD)
        follow_users(driver, usernames_to_follow)
    finally:
        driver.quit()

if __name__ == "__main__":
    main()

```
